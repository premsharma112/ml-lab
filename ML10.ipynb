```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



# Load the dataset
from google.colab import files
uploaded = files.upload()


# Read the uploaded file
df = pd.read_csv(next(iter(uploaded)))
```

```python
print(df.head())

# Bar Chart (Example: Mean Glucose level per Outcome)
plt.figure(figsize=(8, 5))
df.groupby('Outcome')['Glucose'].mean().plot(kind='bar', color=['blue', 'red'])
plt.xlabel("Diabetes Outcome (0 = No, 1 = Yes)")
plt.ylabel("Mean Glucose Level")
plt.title("Mean Glucose Level by Diabetes Outcome")
plt.show()

# Histogram (Example: Distribution of Glucose Levels)
plt.figure(figsize=(8, 5))
sns.histplot(df['Glucose'], bins=20, kde=True, color='purple')
plt.xlabel("Glucose Level")
plt.ylabel("Frequency")
plt.title("Distribution of Glucose Levels")
plt.show()

# Pair Plot (Visualizing relationships among numerical variables)
sns.pairplot(df, hue="Outcome", diag_kind="kde", palette="husl")
plt.show()
```

```python
# Scatter plot with regression line
plt.figure(figsize=(8, 5))
sns.regplot(x=df['Glucose'], y=df['BloodPressure'], scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.xlabel("Glucose Level")
plt.ylabel("Blood Pressure (mm Hg)")
plt.title("Relationship between Glucose and Blood Pressure")
plt.show()
```
